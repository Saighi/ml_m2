```python
from vmdpy import VMD
```

```python
output = VMD(demo_signal,2000,0,8,0,1,1e-7)
```

```python
plt.plot(output[0][2])
```

```python
file_h5 = h5py.File("VMD.h5", 'w')
for key in h5_datas.keys():
    data = []
    for signal in h5_datas[key]:
        data.append(VMD(signal,2000,0,8,0,1,1e-7))
    data = np.array(data)
    for component in data[:,component] : 
        file_h5.create_dataset(key+"_"+str(component), data=data)
file_h5.close()
```
